```python
import pandas as pd
```

```python
x_order_report = pd.read_excel('Company X - Order Report.xlsx')
x_pincode=pd.read_excel('Company X - Pincode Zones.xlsx')
x_sku=pd.read_excel('Company X - SKU Master.xlsx')
courier_company=pd.read_excel("Courier Company - Invoice.xlsx")
c_rates = pd.read_excel('Courier Company - Rates.xlsx')
```

```python
x=pd.merge(x_order_report,x_sku,on="SKU")
x.head()
```

```python
x['Weight (g)'].unique()
```

```python
x.rename(columns={'Weight (g)':'weight_x'},inplace= True)
x.rename(columns={'ExternOrderNo': 'Order ID'}, inplace = True)
x['weight_x']=x['weight_x']/1000

def slab_x(n):
    if n<=0.5:
        return 0.5
    elif n>0.5 and n<=1:
        return 1
x['weight_slab_x']=x['weight_x'].map(slab_x)
x.head()
```

```python
x['weight_slab_x'].max()
```

```python
company = pd.merge(x ,courier_company, on ='Order ID')
company.head()
```

```python
x_pincode.rename(columns={'Zone':'zone_x'},inplace=True)
company.rename(columns={'Zone':"zone_c"},inplace = True)
final=pd.merge(x_pincode,company ,on ="Warehouse Pincode")
final['weight_c']=final['weight_x']
final.head(50)
```

```python
count=0
for i,j in zip (final['zone_x'],final['zone_c']):
    if i!=j:
        count+=1
print(count)
```
